<h2>Spliting features into train_test_split</h2>

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,f1_score,recall_score,precision_score,accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
# Apply StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['cp', 'thal']])
df[['cp', 'thal']] = scaled_features

# Convert target variable into categorical by binning
num_bins = 5
df['target_category'] = pd.cut(df['target'], bins=num_bins, labels=False)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[['cp', 'thal']], df['target_category'], test_size=0.2, random_state=42)
num_bins = 5

```

<h1>Implementing Algorithm</h1>

```python
# Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict(X_test)
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
print("Logistic Regression Accuracy:", accuracy_logreg)

# KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print("KNN Accuracy:", accuracy_knn)

# Single-layer Perceptron
perceptron = MLPClassifier(hidden_layer_sizes=(1,), activation='logistic', max_iter=1000)
perceptron.fit(X_train, y_train)
y_pred_perceptron = perceptron.predict(X_test)
accuracy_perceptron = accuracy_score(y_test, y_pred_perceptron)
print("Single-layer Perceptron Accuracy:", accuracy_perceptron)
```
